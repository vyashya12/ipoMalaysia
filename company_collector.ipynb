```python
from pypdf import PdfReader
# Read from PDF
reader = PdfReader('List_of_Companies.pdf')

# Initialize list to store data
companies = []
final_list = []

# Get number of pages
num_pages = reader.get_num_pages()
print(num_pages)
```

```python
# Append all companies from pdf to final_list array with stock code and needed symbols
final_list = []
for index in range(1, num_pages):
    page = reader.pages[index]
    text = page.extract_text()
    company_entries = text.strip().split('\n')
    if(index == 1):
        for entry in company_entries[1:-1]:
            company_name = entry.split(" ")
            companies.append(company_name[1:-1])

        companies[38] = companies[38][:-5]
        for company in companies:
            entry = " ".join(company)
            final_list.append({"Company": " ".join(company[:-1]), "Stock": company[-1], "YFinCode": company[-1] + ".KL"})
    else:
        for entry in company_entries[1:-1]:
            company_name = entry.split(" ")
            companies.append(company_name[1:-1])
        for company in companies:
            entry = " ".join(company)
            final_list.append({"Company": " ".join(company[:-1]), "Stock": company[-1], "YFinCode": company[-1] + ".KL"})
    companies = []
print(final_list)
```

```python
import pandas as pd

file_path = './data/AllCompanyDataLatest.xlsx'
df = pd.read_excel(file_path)

def process_executives(executives_str):
    if pd.isna(executives_str):
        return []
    executives_list = executives_str.split(';')
    executives = []
    for exec_str in executives_list:
        name_title = exec_str.split(',')
        if len(name_title) == 2:
            name = name_title[0].strip()
            title = name_title[1].strip()
            executives.append({'name': name, 'title': title})
    return executives

df['CompanyExecutives'] = df['CompanyExecutives'].apply(process_executives)


for index, row in df.iterrows():
    # Convert the row to a dictionary
    data = row.to_dict()
    # Insert the dictionary into the MongoDB collection
    details.insert_one(data)

print("Data inserted into MongoDB collection 'companies_details'")
```

```python
import requests
from requests.structures import CaseInsensitiveDict
import pandas as pd
from urllib.parse import quote

csv_file_path = './data/AllCompanyDataLatest.csv'  # Replace with your actual file path
df = pd.read_csv(csv_file_path)

def encode_address(address):
    stringed = str(address)
    return quote(stringed)

df["encoded_address2"] = df["address2"].apply(encode_address)

def get_geocode_data(encoded_address):
    api_key = "60eb608d479f4216ae04f1e325d1ab2f"
    url = f"https://api.geoapify.com/v1/geocode/search?text={encoded_address}&filter=countrycode:MY&apiKey={api_key}"
    headers = CaseInsensitiveDict()
    headers["Accept"] = "application/json"
    
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            json_resp = response.json()
            longitude = json_resp["features"][0]["geometry"]["coordinates"][0]
            latitude = json_resp["features"][0]["geometry"]["coordinates"][1]
            return longitude, latitude
        else:
            return None, None
    except Exception as e:
        return None, None

# Create new columns for longitude and latitude
df['longitude'] = None
df['latitude'] = None

# Iterate over the DataFrame rows and update the longitude and latitude
for index, row in df.iterrows():
    encoded_address = row['encoded_address2']
    longitude, latitude = get_geocode_data(encoded_address)
    df.at[index, 'longitude'] = longitude
    df.at[index, 'latitude'] = latitude

# Show the DataFrame with the new columns
print(df)

# Optionally, save the DataFrame back to a CSV file
output_csv_file_path = './data/CompanyWithLongLat.csv'  # Replace with your desired output file path
df.to_csv(output_csv_file_path, index=False)
```
